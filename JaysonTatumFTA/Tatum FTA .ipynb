```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from nba_api.stats.static import teams 
from nba_api.stats.endpoints.leaguedashplayerstats import LeagueDashPlayerStats
```

```python
pd.set_option('display.max_colwidth',500)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',500)
```

```python
nba_teams = teams.get_teams()
celtics = [team for team in nba_teams if team['city']=='Boston']
celtics_id = celtics[0].get('id')
print('Boston Celtics ID: {}'.format(celtics_id))
```

```python
total_stats = LeagueDashPlayerStats(per_mode_detailed='PerGame').get_data_frames()[0]
print(len(total_stats))
total_stats
```

```python
filtered_stats = total_stats[(total_stats.MIN>=20.0) & (total_stats.GP>=20)].reset_index()
player_ids = filtered_stats['PLAYER_ID']

# Month Parameter: December=3, January=4, February=5 etc. 
total_stats_per100 = LeagueDashPlayerStats(per_mode_detailed='Per100Possessions').get_data_frames()[0]
total_stats_per100 = total_stats_per100[total_stats_per100.PLAYER_ID.isin(player_ids)].reset_index()
total_stats_per100
```

```python
fta_per100 = total_stats_per100[['PLAYER_ID', 'PLAYER_NAME', 'TEAM_ID', 'FTM', 'FTA', 'FT_PCT']]
fta_per75 = fta_per100.copy()

for cat in ['FTM', 'FTA']:
    fta_per75[cat] = fta_per75[cat].map(lambda num: num*0.75)
    
average_fta_per75 = fta_per75['FTA'].mean()
print('Average FTA per 75 Possesions: {}'.format(average_fta_per75))
fta_per75
```

```python
plt.style.use('seaborn')

plt.hist(fta_per75.FTA, bins=75)

plt.axvline(x=average_fta_per75, color='r', linestyle='-', linewidth=3)
plt.text(x=average_fta_per75+0.13, y=13, s='Average: {:0.2f} FTA'.format(average_fta_per75), fontsize=12, fontweight='bold')

plt.title('FTA per 75 Posessions', fontweight='bold', fontsize=17)
plt.xlabel('FTA (per 75 Possessions)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.show()
```

```python
#tatum = fta_per75[fta_per75.PLAYER_NAME=='Jayson Tatum']

celtics_stats = []
for index, row in fta_per75.iterrows():
    if row.TEAM_ID == celtics_id: 
        if row.PLAYER_NAME != 'Al Horford' and row.PLAYER_NAME != 'Moses Brown':
            celtics_stats.append((row.PLAYER_NAME, row.FTA, row.FT_PCT))
    elif row.PLAYER_NAME == 'Kemba Walker':
        celtics_stats.append((row.PLAYER_NAME, row.FTA, row.FT_PCT))

celtics_last_names = [player[0].split()[-1] for player in celtics_stats]
celtics_stats
```

```python
plt.scatter(fta_per75.FTA, fta_per75.FT_PCT, color='dimgrey', edgecolor='black', alpha=0.35)
plt.scatter(x=[player[1] for player in celtics_stats], y=[player[2] for player in celtics_stats], color='tomato', edgecolor='black', s=75, label='Min. 20 MPG, 20 GP')
for player in celtics_stats: 
    if player[0] == 'Jayson Tatum':
        plt.text(x=player[1]+0.07, y=player[2]+0.007, s='{}'.format(player[0]), fontsize=11, fontweight='bold')
    else: 
        plt.text(x=player[1]+0.07, y=player[2]+0.007, s='{}'.format(player[0]), fontsize=11)

plt.axvline(x=average_fta_per75, color='black', linestyle='--', linewidth=2)
plt.text(x=average_fta_per75+0.13, y=0.98, s='Average: {:0.2f} FTA'.format(average_fta_per75), fontsize=13, fontweight='bold')

plt.title('FTA per 75 Posessions\n(2020-21 Boston Celtics)', fontweight='bold', fontsize=17)
plt.xlabel('FTA (per 75 Possessions)', fontsize=14)
plt.ylabel('FT PCT (%)', fontsize=14)

plt.legend(loc=4, fontsize='large')
plt.show()
```

```python
all_nba = ['Joel Embiid', 'Giannis Antetokounmpo', 'Paul George', 'Stephen Curry', 'Kawhi Leonard', 'Nikola Jokic', 
             'Luka Doncic', 'LeBron James', 'Julius Randle', 'Damian Lillard', 'Chris Paul', 'Jimmy Butler', 
             'Rudy Gobert', 'Bradley Beal', 'Kyrie Irving', 'Anthony Davis', 'James Harden', 'Pascal Siakam',
             'Ben Simmons', 'Russell Westbrook', 'Kevin Durant', 'Blake Griffin', 'Kemba Walker']

all_nba_stats = []
tatum_stats = []
for index, row in fta_per75.iterrows():
    if row.PLAYER_NAME in all_nba: 
        all_nba_stats.append((row.PLAYER_NAME, row.FTA, row.FT_PCT))
    if row.PLAYER_NAME == 'Jayson Tatum':
        tatum_stats.append((row.PLAYER_NAME, row.FTA, row.FT_PCT))
        tatum_id = row.PLAYER_ID

all_nba_sum = 0
for i in range(len(all_nba_stats)):
    all_nba_sum += all_nba_stats[i][1]
all_nba_sum += celtics_stats[2][1]
all_nba_average = all_nba_sum/(len(all_nba_stats)+1)

print(all_nba_average)
all_nba_stats
```

```python
plt.scatter(fta_per75.FTA, fta_per75.FT_PCT, color='dimgrey', edgecolor='black', alpha=0.35)
plt.scatter(x=[player[1] for player in all_nba_stats], y=[player[2] for player in all_nba_stats], color='gold', edgecolor='black', s=75, label='All-NBA Team (2019-21)')
for player in all_nba_stats: 
    plt.text(x=player[1]+0.07, y=player[2]+0.007, s='{}'.format(player[0].split()[-1]), fontsize=11)
plt.scatter(x=tatum_stats[0][1], y=tatum_stats[0][2], color='tomato', edgecolor='black', s=75, label='Jayson Tatum')
#plt.text(x=tatum_stats[0][1]+0.07, y=tatum_stats[0][2]+0.007, s='Tatum', fontweight='bold')

plt.axvline(x=average_fta_per75, color='black', linestyle='--', linewidth=2, label='League Average (3.3 FTA)')
#plt.text(x=average_fta_per75+0.13, y=0.98, s='Average: {:0.2f} FTA'.format(average_fta_per75), fontsize=13, fontweight='bold')
plt.axvline(x=tatum_stats[0][1], color='tomato', linestyle='--', linewidth=2)
plt.axvline(x=all_nba_average, color='goldenrod', linestyle='--', linewidth=2, label='All-NBA Average (6.3 FTA)')

plt.title('FTA per 75 Posessions\n(2019-21 All-NBA Team)', fontweight='bold', fontsize=17)
plt.xlabel('FTA (per 75 Possessions)', fontsize=14)
plt.ylabel('FT PCT (%)', fontsize=14)

plt.legend(fontsize='large', loc=4)
plt.show()
```

```python
months = ['December', 'January', 'February', 'March', 'April', 'May']
tatum_per_month = []
all_nba_average_per_month = []
nba_average_per_month = []


count=0
for month in range(3,9):
    per_month_stats = LeagueDashPlayerStats(per_mode_detailed='Per100Possessions', month=month).get_data_frames()[0]
    per_month_stats = per_month_stats[per_month_stats.PLAYER_ID.isin(player_ids)].reset_index()
    all_nba_stats = per_month_stats[per_month_stats.PLAYER_NAME.isin(all_nba)].reset_index()
    
    for index, row in per_month_stats.iterrows():
        if row.PLAYER_NAME == 'Jayson Tatum':
            tatum_per_month.append((months[count], row.GP, row.FTA*0.75, row.FT_PCT, row.PTS*0.75))
            
    nba_average_per_month.append((months[count], per_month_stats.FTA.mean()*0.75))
    all_nba_average_per_month.append((months[count], all_nba_stats.FTA.mean()*0.75, all_nba_stats.FT_PCT.mean()*0.75, 
                                      all_nba_stats.PTS.mean()*0.75))
    count += 1
            
tatum_per_month
#all_nba_average_per_month
#nba_average_per_month
```

```python
nba_average_fta = [month[1] for month in nba_average_per_month]
all_nba_average_fta = [month[1] for month in all_nba_average_per_month]

tatum_gp = [month[1]*10 for month in tatum_per_month]
tatum_fta = [month[2] for month in tatum_per_month]
tatum_ft_pct = [month[3] for month in tatum_per_month]
tatum_pts = [month[4] for month in tatum_per_month]

all_nba_average_pct_pts_off_ft = []
tatum_pct_pts_off_ft = []
for i in range(len(tatum_fta)):
    tatum_pct = (tatum_ft_pct[i]*tatum_fta[i])/tatum_pts[i]
    tatum_pct_pts_off_ft.append(tatum_pct)
    all_nba_pct = ((all_nba_average_per_month[i][1]*all_nba_average_per_month[i][2])/all_nba_average_per_month[i][3])
    all_nba_average_pct_pts_off_ft.append(all_nba_pct)
tatum_pct_pts_off_ft = np.array(tatum_pct_pts_off_ft)
all_nba_average_pct_pts_off_ft = np.array(all_nba_average_pct_pts_off_ft)
print(tatum_pct_pts_off_ft)
print(all_nba_average_pct_pts_off_ft)
```

```python
plt.scatter(x=months, y=tatum_fta, s=tatum_pct_pts_off_ft*750, color='tomato', edgecolor='black', label='Jayson Tatum')
plt.plot(months, tatum_fta, color='indianred', linestyle='dotted', alpha=0.5)

plt.scatter(x=months, y=all_nba_average_fta, s=all_nba_average_pct_pts_off_ft*750, color='gold', edgecolor='black',
           label='All-NBA Team (2019-21)')
plt.plot(months, all_nba_average_fta, color='goldenrod', linestyle='dotted', alpha=0.5)

plt.plot(months, nba_average_fta, c='midnightblue', alpha=0.75, linewidth=2, linestyle='solid', 
         label='League Average (Min. 20 MPG, 20 GP)')
for i in range(len(tatum_fta)):
    plt.text(x=months[i], y=tatum_fta[i]+0.13, s='{:0.2f}'.format(tatum_fta[i]))
    plt.text(x=months[i], y=all_nba_average_fta[i]+0.13, s='{:0.2f}'.format(all_nba_average_fta[i]))
    plt.text(x=months[i], y=nba_average_fta[i]+0.1, s='{:0.2f}'.format(nba_average_fta[i]))

plt.title('FTA per 75 Possessions by Month', fontsize=17, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('FTA (per 75 Possessions)', fontsize=14)

plt.legend()
plt.show()
```
